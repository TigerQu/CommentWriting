```python
# external resources utilized: Google Gemini 3 Flash Preview
# work distribution: Tiger: Comment Generator Carlie: File I/O
# PLEASE USE ANACONDA KERNEL!!!
import csv
from google import genai

gemini_api_key = 'AIzaSyDWTZR1lDCCcT7NR02eC80ReKSpGwqEo10'
client = genai.Client(api_key=gemini_api_key)


def generate_comment(name, grade, cn, ce, st, ag, excerpt):
    '''This function takes the input of the student's
    information and call Gemini to generate a comment.'''
    system_instruction = f'''You are now writing a comment
            for a student's
            semester report. **Your output should only
            include the comment.** You
            will be given the course description, grade
            of the student, strengths,
            and areas of growth. Use encouraging words and
            second person POV.
            Not to mention grade explicitly. You
            can adopt some content from student's essay
            excerpt. Here is an example:
            This semester in AP AB Calculus, we covered
            the topics of limits, continuity,
            and differentiation and began the study of
            integration, which we will continue
            into the second semester. In addition to
            daily homework and frequent quizzes,
            we had three in-class unit tests, a cumulative
            final exam, and two projects,
            one on predicting stock prices using derivatives
            and one on modeling fluid
            flow using related rates and Torricelli’s Law.
            \n Migo, you have a stellar
            work ethic and student habits. You faithfully
            complete the nightly homework
            and are fully engaged in class discussions.
            You are not really comfortable
            asking questions during the larger class discussion
            but you work well in
            small groups with your peers. Migo, you have not
            needed to come to see me
            outside of class for extra help, and you seem to
            grasp the content easily.
            \n In addition, it is better to list the strengths
            and areas of growth,
            then follow by a paragraph of summary. The following
            part is the detailed
            info about the corresponding course and student
            detail. Student Attributes:
            {name} \n Student Grade: {grade} \n Course Name:
            {cn} \n Course
            Expectations: {ce} \n strengths: {st} \n areas of growth:
            {ag} \n student
            essay excerpt: {excerpt} (The excerpt may not exist)
            You may reorganize
            the language, but you must follow the truth and
            surpress hallucination.
            Now generate the comment for the report.'''
    print(f"Generating comment for {name}")
    response = client.models.generate_content(model='gemini-3-flash-preview',
                                              contents=system_instruction)
    print(f'Generated comment for {name}')
    print(response.text)
    return response.text


def new_empty():
    '''This function generates an
    empty csv file for teachers'''
    with open('teacher.csv', 'w') as csvfile:
        myfile = csv.writer(csvfile)
    myfile.writerow([
        "student name", "student grade", "course name", "course expectations",
        "strengths", "areas of growth", "essay excerpts"
    ])
    return


def file_io(if_newfile: bool):
    '''This fuction handles the file io'''
    if if_newfile:
        new_empty()
    with open('teacher.csv') as csvfile:
        data = csv.reader(csvfile, delimiter=',')
        header = next(data)
        for row in data:
            filename = f'{row[0]}.txt'
            res = generate_comment(row[0], row[1], row[2], row[3], row[4],
                                   row[5], row[6])
            fout = open(filename, 'w')
            fout.write(res)


file_io(False)
```
